## Just GaN

```python
GaN = mat_data.GaN(doping_type='n', doping=0) # Pre define materials

list_layers = [(1, GaN)] # Create a list of our layers
layers = [material.layer(*el) for el in list_layers]

device = band.band_diagram(layers=layers) # Create the Band diagram

device.bend() # Calculate the band diagram
fig = device.plot(display_E0=False) # Display_E0 is False by default
fig.update_yaxes(autorange=False, range=[-2,2])
fig.show(config=my_config)
```

## Schottky Diode GaN

```python
GaN = mat_data.GaN(doping_type='n', doping=1e16) # Pre define materials
Ni = material.metal(work_function=5.35) # up to 5.35

list_layers = [(0.5, Ni), (1, GaN)] # Create a list of our layers
layers = [material.layer(*el) for el in list_layers]

device = band.band_diagram(layers=layers) # Create the Band diagram

device.bend() # Calculate the band diagram
fig = device.plot(display_E0=False) # Display_E0 is False by default
fig.show(config=my_config)
```
